```python
for page in range(1,51,1) :
    url = 'https://movie.naver.com/movie/bi/mi/point.naver?code=106360'+str(page)
    res = req.get(url)
    soup = bs(res.text,'lxml')
    
    reples = soup.select('div.score_reple > p')
    
    icon = soup.select('span_ico_viewer')
    
    for index in icon :
        index.extract()
        
    for reple in reples :
        print(reple.text.strip())
```

```python
# range() : 일정한 간격으로 숫자를 증가시킬 수 있는 함수
```

```python
from tqdm import tqdm_notebook as tqdm
```

```python
for page in tqdm(range(1,51,1)) :
    url = 'https://movie.naver.com/movie/bi/mi/point.naver?code=106360'+str(page)
    res = req.get(url)
    soup = bs(res.text,'lxml')
    
    reples = soup.select('div.score_reple > p')
    
    icon = soup.select('span_ico_viewer')
    
    for index in icon :
        index.extract()
        
    for reple in reples :
        print(reple.text.strip())
```
